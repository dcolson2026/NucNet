```python
def mass_num_vs_abundance_lists(filename):
    file = open(filename, "r")
    data1 = file.readlines()
    data_list1 = []

    for line in data1:
        if line[0] == "#":
            continue
        line = line.strip()
        data_list1.append(line.split(" "))
    cut = data_list1[2:]
    mass_number_list = [int(i[0]) for i in cut]
    abundance_list = [float(i[2]) for i in cut]
    
    return mass_number_list, abundance_list
```

### 32. Comparing network calculations to equilibrium


```python
""" When all nuclear reactions occur rapidly compared to the timescale on which a system 
changes, the system can achieve an equilibrium (a free energy minimum or entropy maximum). 
The most relevant of these for our present purposes is nuclear statistical equilibrium (NSE), 
which occurs when all nuclear reactions except weak nuclear reactions are sufficiently fast. 
A second important equilibrium is quasi-statistical equilibrium (QSE), 
which is nuclear statistical equilibrium with the added constraint that the reactions 
that change the number of heavy nuclei are too slow. I will compare the abundances in my 
r-process calculation to the abundances in those two equilibria to understand how the system 
evolves. """
"""Well approximated by NSE early then QSE later"""
```
